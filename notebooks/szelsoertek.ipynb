```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
def f1(x,y):
    return x**2 + y**2

def f3(x,y):
    return x*y

def f2(x,y):
    return 1 - x**2 - y**2
```

```python
space = np.linspace(-10,10,100)

X,Y = np.meshgrid(space,space)
```

```python
Z1 = f1(X,Y)

Z2 = f2(X,Y)

Z3 = f3(X,Y)
```

```python
fig = plt.figure(3)


first =fig.add_subplot(131, projection='3d').plot_surface(X, Y, Z1, cmap='viridis')
fig.add_subplot(132, projection='3d').plot_surface(X, Y, Z2, cmap='plasma')
fig.add_subplot(133, projection='3d').plot_surface(X, Y, Z3, cmap='inferno')
plt.show()
```
