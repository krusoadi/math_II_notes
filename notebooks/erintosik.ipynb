```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
space = np.linspace(-10,10,100)

X,Y = np.meshgrid(space,space)

sik_space = np.linspace(-5,5,100)

X1, Y1 = np.meshgrid(sik_space,sik_space)
```

```python
def f(x,y):
    return x**2 +  y**2

def f2x(x,y):
    return 2*x**2 +  y**2

def f2y(x,y):
    return x**2 + 2*y**2

def erintosik(x,y,x0,y0,z0):
    return f(x0,y0) + f2x(x0,y0)*(x-x0) + f2y(x0,y0)*(y-y0)
    
```

```python
Z = f(X, Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, erintosik(X1,Y1,3,3,3), color='red')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('f(x, y)')

ax.view_init(elev=25, azim=30)

# Show the plot
plt.show()
```
